```python
import pandas as pd
import numpy as np
```

false if not null or true if null

```python
d ={'first-score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(d)

mv = df.isnull()
print(mv)
```

```python
df.fillna(0)
```

```python
df.fillna(method='pad')
```

```python
df.fillna(method='bfill')
```

use of interpolate() fills  missing value with techniques like linear method

```python
import pandas as pd

df = pd.DataFrame({"A": [12, 4, 5, None, 1],
                   "B": [None, 2, 54, 3, None],
                   "C": [20, 16, None, 3, 8],
                   "D": [14, 3, None, None, 6]})
print(df)
```

```python
df.interpolate(method ='linear', limit_direction ='forward')
```
