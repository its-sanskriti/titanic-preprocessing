Step 1 â€“ Load Iris Dataset

```python

from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

Step 2 â€“ Apply Scalers

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

# Numeric features (all columns here)
numeric_df = df.copy()

# MinMax
minmax = MinMaxScaler()
df_minmax = pd.DataFrame(minmax.fit_transform(numeric_df), columns=numeric_df.columns)

# Standardization
standard = StandardScaler()
df_standard = pd.DataFrame(standard.fit_transform(numeric_df), columns=numeric_df.columns)

# Robust
robust = RobustScaler()
df_robust = pd.DataFrame(robust.fit_transform(numeric_df), columns=numeric_df.columns)
```

Step 3 â€“ Visualize the Scaling

```python
import matplotlib.pyplot as plt

features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

fig, axes = plt.subplots(4, 4, figsize=(16,12))
scalers = {'Original': numeric_df, 'MinMax': df_minmax, 'Standard': df_standard, 'Robust': df_robust}

for i, feature in enumerate(features):
    for j, (name, data) in enumerate(scalers.items()):
        axes[i, j].hist(data[feature], bins=20)
        if i == 0:
            axes[i, j].set_title(name)
        if j == 0:
            axes[i, j].set_ylabel(feature)

plt.tight_layout()
plt.show()

```
